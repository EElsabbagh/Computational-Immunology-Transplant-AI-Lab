```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import entropy

# 1. Create a tiny synthetic TCR dataset
tcr = pd.DataFrame({
    "clone_id": [f"clone_{i}" for i in range(1, 51)],
    "clone_count": np.random.randint(1, 100, size=50)
})

tcr.head()
```

```python
counts = tcr['clone_count'].values
shannon = entropy(counts)
clonality = 1 - shannon / np.log(len(counts))

print("Shannon diversity:", shannon)
print("Clonality index:", clonality)
```

```python
plt.hist(counts, bins=10)
plt.xlabel("Clone size")
plt.ylabel("Frequency")
plt.title("TCR Clone Size Distribution (synthetic)")
plt.show()
```
